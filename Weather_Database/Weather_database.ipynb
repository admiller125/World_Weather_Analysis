```python
#import random function
import random

#import citipy 
from citipy import citipy

#import api key 
from config import weather_api_key

import requests

#iniitalize latitude and longitude
lat = [0]*2000
lon = [0]*2000

#generate 2000 random latitudes and longitudes
for i, val in enumerate(lat):
    lat[i] = random.randint(-90,89) + random.random()
    lon[i] = random.randint(-90,89) + random.random()
```

```python
#generate a list of coordinates to iterate over
citi_coord = zip(lat,lon)
citi_coord = list(citi_coord)
```

```python
#convert the coordinates to city and country codes using citipy
cities = [0]*2000
for i, coord in enumerate(citi_coord):
     cities[i] = citipy.nearest_city(coord[0],coord[1]).city_name
```

```python
city_data = []
#query open weather map for all city data
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
#importing time
import time
for i, city in enumerate(cities):
    #counting to ensure number of querys per minute allowed isnt exceeded
    if (i % 50 == 0 and i > 1):
        time.sleep(60)
    #Create URL with each city.
    city_url = url + "&q=" + city
    try:
    #get data for cities
        city_weather = requests.get(city_url).json()
        city_country = city_weather["sys"]["country"]
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_description = city_weather["weather"][0]["description"]
    
    except:
        print("City not found. Skipping...")
        pass
    
    # Indicate that Data Loading is complete.
    print("-----------------------------")
    print("Data Retrieval Complete      ")
    print("-----------------------------")
    
    #create list of dictionaries for cities
    city_data.append({"City": city.title(),
                    "Country": city_country,
                    "Lat": city_lat,
                    "Lng": city_lng,
                    "Max Temp": city_max_temp,
                    "Humidity": city_humidity,
                    "Cloudiness": city_clouds,
                    "Wind Speed": city_wind,
                    "Current Description": city_description})
```

```python
#import pandas and convert to dataframe
import pandas as pd
WeatherPy_DF = pd.DataFrame(city_data)
```

```python
#convert to farenhight
WeatherPy_DF["Max Temp"] = 1.8*(WeatherPy_DF["Max Temp"] - 273.15) + 32 
```
